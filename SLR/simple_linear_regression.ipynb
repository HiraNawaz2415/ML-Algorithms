# Simple Linear Regression
Simple linear regression in machine learning is a basic technique used to model the relationship between one independent variable (feature) and one dependent variable (target/output) by fitting a straight line to the data.

## What is it?
It assumes a linear relationship between the input variable 
x and output variable ùë¶.

- The goal is to find the best-fitting line that passes through by each point  described by the equation:
        **ùë¶=ùëöùë•+ùëè**

where:

y = predicted output (dependent variable),
x = input feature (independent variable),
m = slope of the line (how much 
y changes for a unit change in x
b = y-intercept (value of y when ùë•=0)




## How does it work in ML?
Given training data of pairs 
(ùë•ùëñ,ùë¶ùëñ) the model tries to find 
ùëö and ùëè that minimize the difference between predicted y values and actual ùë¶ values.

This difference is typically measured by the Mean Squared Error (MSE).

The process of finding the best ùëö and b is called fitting the model or training.

- Height on the X-axis,

- Weight on the Y-axis.

 **Weight=ùëö√óHeight+ùëè**

Where:

b = intercept (where the line hits the Y-axis if height = 0).
